```python
# get file name and path of file name
import os
path = "./Input/"
if (os.path.exists(path) == False):
    exit(0)
file = os.listdir(path)
```

```python
# get file name
for i in file:
    print(i)
```

```python
# get path of file name
for name in file:
    print(os.path.join(path, name))
```

```python
# get full path
path = "./Input"
for name in file:
    print(os.path.abspath(os.path.join(path, name)))
```

```python
# get path of file name
import glob
path = "../*"
name = glob.glob(path)
for i in name:
    print(i)
```
