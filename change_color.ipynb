```python
import cv2 # for image
import os #for get path image
import matplotlib.pyplot as plt
```

```python
input_path = "./Input"
output_path = "./Output"
```

```python
# Check path input and output have already exist 
if os.path.exists(input_path) == False:
    os.makedirs(input_path)
elif os.path.exists(output_path) == False:
    os.makedirs(output_path)
```

```python
# Change Color image [BGR](default image of opencv) to grayscale
for filename in os.listdir(input_path):
    path = os.path.join(input_path, filename)
    image = cv2.imread(path)
    if image is not None:
        output_picture = os.path.join(output_path, filename)
        img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        cv2.imwrite(output_picture, img_gray)
```

```python
# Display Image
img = cv2.imread("./Input/1.jpg", 1)
plt.imshow(img)
plt.axis('off')
plt.show()
```
