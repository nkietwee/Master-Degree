```python

# Load the original image
image = cv2.imread("./Input-test/chang_thh.jpg")  # Replace with your image path
original = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 1: Convert to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 2: Apply Thresholding
_, thresholded = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# Step 3: Mask the Cap Area (Optional - Enhance Cap Detection)
# Using morphological operations to isolate the cap
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
masked = cv2.morphologyEx(thresholded, cv2.MORPH_CLOSE, kernel)

# Step 4: Contour Detection to Highlight the Cap
contours, _ = cv2.findContours(masked, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
highlighted = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)  # Convert back to BGR for contour overlay
cv2.drawContours(highlighted, contours, -1, (0, 0, 255), 2)

# Display Results
titles = ["Original", "Grayscale", "Thresholded", "Masked", "Highlighted"]
images = [original, gray, thresholded, masked, highlighted]

plt.figure(figsize=(12, 8))
for i in range(5):
    plt.subplot(2, 3, i+1)
    if i == 0:  # Original is RGB
        plt.imshow(images[i])
    else:  # Others are grayscale
        plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("./Input-test/chang_thh.jpg", cv2.IMREAD_GRAYSCALE)  # Replace with your image path

# Step 1: Thresholding the image
_, binary = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)

# Step 2: Calculate the horizontal projection (sum of pixel values along columns)
horizontal_projection = np.sum(binary, axis=1)

# Step 3: Normalize the histogram for better visualization
normalized_projection = (horizontal_projection / max(horizontal_projection)) * binary.shape[1]

# Step 4: Find regions with significant text content
text_lines = np.where(horizontal_projection > 0)[0]
start_indices = [text_lines[0]]
end_indices = []

for i in range(1, len(text_lines)):
    if text_lines[i] - text_lines[i - 1] > 1:  # Check for a gap
        end_indices.append(text_lines[i - 1])
        start_indices.append(text_lines[i])

end_indices.append(text_lines[-1])

# Draw horizontal lines to highlight text regions
segmented_image = cv2.cvtColor(binary, cv2.COLOR_GRAY2BGR)
for start, end in zip(start_indices, end_indices):
    cv2.line(segmented_image, (0, start), (binary.shape[1], start), (0, 0, 255), 2)
    cv2.line(segmented_image, (0, end), (binary.shape[1], end), (0, 0, 255), 2)

# Step 5: Plot results
plt.figure(figsize=(12, 6))

# Original segmented image
plt.subplot(1, 2, 1)
plt.imshow(segmented_image, cmap="gray")
plt.title("Text Segmentation with Projection")
plt.axis("off")

# Horizontal projection
plt.subplot(1, 2, 2)
plt.plot(horizontal_projection, np.arange(binary.shape[0]), color="black")
plt.gca().invert_yaxis()  # Invert y-axis to align with the image
plt.fill_betweenx(np.arange(binary.shape[0]), horizontal_projection, color="black")
plt.title("Horizontal Projection")
plt.xlabel("Sum of pixel values")
plt.ylabel("Image rows")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import pytesseract
from pytesseract import Output

# Configure Tesseract path (update with your installation path)
pytesseract.pytesseract.tesseract_cmd = r"/bin/tesseract"

def preprocess_image(image_path):
    """Preprocess the image."""
    # Step 1: Load the image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Step 2: Apply Gaussian Blur
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Step 3: Increase contrast
    contrast = cv2.addWeighted(blurred, 2, gray, -1, 0)

    # Step 4: Apply sharpening
    kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
    sharpened = cv2.filter2D(contrast, -1, kernel)

    # Step 5: Binary thresholding
    _, binary = cv2.threshold(sharpened, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # Step 6: Hough Transform to detect circular features (optional for bottle caps)
    circles = cv2.HoughCircles(binary, cv2.HOUGH_GRADIENT, dp=1.2, minDist=50, param1=50, param2=30, minRadius=30, maxRadius=200)
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for circle in circles[0, :]:
            center = (circle[0], circle[1])
            radius = circle[2]
            cv2.circle(image, center, radius, (0, 255, 0), 2)

    return binary, image

def detect_text_regions(binary_image):
    """Detect text regions using contour detection."""
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    text_regions = []
    
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        # Filter small boxes
        if w > 20 and h > 10:
            text_regions.append((x, y, w, h))
    
    return text_regions

def extract_text(image, text_regions):
    """Extract text using Tesseract OCR."""
    results = []
    for (x, y, w, h) in text_regions:
        roi = image[y:y+h, x:x+w]
        text = pytesseract.image_to_string(roi, config="--psm 6")
        results.append((text, (x, y, w, h)))
    return results

# Main Pipeline
image_path = "./Input-test/Nestle.jpg"  # Update with your image path

# Step 1: Preprocess Image
binary_image, preprocessed_image = preprocess_image(image_path)

# Step 2: Detect Text Regions
text_regions = detect_text_regions(binary_image)

# Step 3: Extract Text
image = cv2.imread(image_path)
detected_text = extract_text(image, text_regions)

# Visualize Results
for text, (x, y, w, h) in detected_text:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(image, text.strip(), (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Display the processed image
cv2.imshow("Detected Text", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
